```python
# Plot altitude over time (ms)
plt.figure(figsize=(10, 6))
plt.plot(df['ms'], df['alt'], label='Altitude', color='b')
plt.title('Altitude over Time')
plt.xlabel('Time (ms)')
plt.ylabel('Altitude (alt)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df['total_acceleration'] = np.sqrt(df['aX']**2 + df['aY']**2 + df['aZ']**2)

# Plot the total acceleration over time (ms)
plt.figure(figsize=(10, 6))
plt.plot(df['ms'], df['total_acceleration'], label='Total Acceleration', color='r')
plt.title('Total Acceleration over Time')
plt.xlabel('Time (ms)')
plt.ylabel('Total Acceleration (m/s²)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate the derivative (rate of change) of acceleration for aX, aY, and aZ
df['aX_derivative'] = np.gradient(df['aX'], df['ms'])
df['aY_derivative'] = np.gradient(df['aY'], df['ms'])
df['aZ_derivative'] = np.gradient(df['aZ'], df['ms'])

# Plot the derivatives of aX, aY, and aZ over time (ms)
plt.figure(figsize=(10, 6))

plt.plot(df['ms'], df['aX_derivative'], label='aX Derivative', color='b')
plt.plot(df['ms'], df['aY_derivative'], label='aY Derivative', color='g')
plt.plot(df['ms'], df['aZ_derivative'], label='aZ Derivative', color='r')

plt.title('Acceleration Derivatives over Time')
plt.xlabel('Time (ms)')
plt.ylabel('Rate of Change of Acceleration (m/s² per ms)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Calculate total acceleration (magnitude of the vector formed by aX, aY, and aZ)
df['total_acceleration'] = np.sqrt(df['aX']**2 + df['aY']**2 + df['aZ']**2)

# Rate of change of altitude (difference between consecutive points)
df['alt_derivative'] = np.gradient(df['alt'], df['ms'])

# Threshold for rapid changes (2 standard deviations above the mean)
alt_change_threshold = df['alt_derivative'].std() * 2
rapid_alt_changes = df[np.abs(df['alt_derivative']) > alt_change_threshold]

# Rate of change of total acceleration (difference between consecutive points)
df['total_acceleration_derivative'] = np.gradient(df['total_acceleration'], df['ms'])

# Threshold for rapid acceleration changes
acc_change_threshold = df['total_acceleration_derivative'].std() * 2
rapid_acc_changes = df[np.abs(df['total_acceleration_derivative']) > acc_change_threshold]

# Plot both altitude and total acceleration with points of rapid change highlighted
plt.figure(figsize=(12, 8))

# Plot altitude over time
plt.subplot(2, 1, 1)
plt.plot(df['ms'], df['alt'], label='Altitude (Alt)', color='b')
plt.scatter(rapid_alt_changes['ms'], rapid_alt_changes['alt'], color='r', label='Rapid Altitude Change', zorder=5)
plt.title('Altitude Over Time with Rapid Changes Highlighted')
plt.xlabel('Time (ms)')
plt.ylabel('Altitude (m)')
plt.legend()
plt.grid(True)

# Plot total acceleration over time
plt.subplot(2, 1, 2)
plt.plot(df['ms'], df['total_acceleration'], label='Total Acceleration', color='g')
plt.scatter(rapid_acc_changes['ms'], rapid_acc_changes['total_acceleration'], color='r', label='Rapid Acceleration Change', zorder=5)
plt.title('Total Acceleration Over Time with Rapid Changes Highlighted')
plt.xlabel('Time (ms)')
plt.ylabel('Total Acceleration (m/s²)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
pressure_acceleration_corr = df[['pressure', 'aX', 'aY', 'aZ']].corr()

# Now let's visualize the relationship between pressure and each acceleration component using scatter plots
plt.figure(figsize=(15, 5))

# Pressure vs aX (Horizontal Acceleration)
plt.subplot(1, 3, 1)
plt.scatter(df['pressure'], df['aX'], alpha=0.5, color='b')
plt.title('Pressure vs aX (Horizontal Acceleration)')
plt.xlabel('Pressure (Pa)')
plt.ylabel('aX (m/s²)')

# Pressure vs aY (Lateral Acceleration)
plt.subplot(1, 3, 2)
plt.scatter(df['pressure'], df['aY'], alpha=0.5, color='g')
plt.title('Pressure vs aY (Lateral Acceleration)')
plt.xlabel('Pressure (Pa)')
plt.ylabel('aY (m/s²)')

# Pressure vs aZ (Vertical Acceleration)
plt.subplot(1, 3, 3)
plt.scatter(df['pressure'], df['aZ'], alpha=0.5, color='r')
plt.title('Pressure vs aZ (Vertical Acceleration)')
plt.xlabel('Pressure (Pa)')
plt.ylabel('aZ (m/s²)')

plt.tight_layout()
plt.show()

# Display correlation matrix for pressure and acceleration components
print(pressure_acceleration_corr)
```

```python
# Identifying the times for the clusters in the Pressure vs aX plot
# Cluster 1: Points where aX is near zero (around stable flight)
cluster_1 = df[(df['aX'] > -1) & (df['aX'] < 1)]

# Cluster 2: Points where aX decreases to around -10 m/s² (possible aerodynamic forces)
cluster_2 = df[(df['aX'] < -9) & (df['aX'] > -11)]

# Now let's extract the time values (ms) for these clusters
cluster_1_times = cluster_1['ms']
cluster_2_times = cluster_2['ms']

# Displaying the time ranges for both clusters
cluster_1_time_range = (cluster_1_times.min(), cluster_1_times.max())
cluster_2_time_range = (cluster_2_times.min(), cluster_2_times.max())

# Output time ranges
print(f"Cluster 1 (aX near zero) time range: {cluster_1_time_range}")
print(f"Cluster 2 (aX around -10 m/s²) time range: {cluster_2_time_range}")
```

```python
# Identifying the clusters for aY (lateral acceleration)
# Cluster 1: Points where aY is near zero (around stable flight)
cluster_1_aY = df[(df['aY'] > -1) & (df['aY'] < 1)]

# Cluster 2: Points where aY decreases to around -10 m/s² (possible lateral aerodynamic forces)
cluster_2_aY = df[(df['aY'] < -9) & (df['aY'] > -11)]

# Now let's extract the time values (ms) for these clusters
cluster_1_aY_times = cluster_1_aY['ms']
cluster_2_aY_times = cluster_2_aY['ms']

# Displaying the time ranges for both clusters
cluster_1_aY_time_range = (cluster_1_aY_times.min(), cluster_1_aY_times.max())
cluster_2_aY_time_range = (cluster_2_aY_times.min(), cluster_2_aY_times.max())

# Output time ranges
print(f"Cluster 1 (aY near zero) time range: {cluster_1_aY_time_range}")
print(f"Cluster 2 (aY around -10 m/s²) time range: {cluster_2_aY_time_range}")
```
