```python
import pandas as pd

data = pd.read_csv('GoannaFinal105per 1.csv')
data.head()
```

```python
columns_to_keep = [
    'Time (s)', 'Altitude (m)', 'Vertical velocity (m/s)', 'Vertical acceleration (m/s²)',
    'Total velocity (m/s)', 'Total acceleration (m/s²)', 'Position East of launch (m)', 
    'Position North of launch (m)', 'Pitch rate (rad/s)', 'Yaw rate (rad/s)'
]

filtered_data = data[columns_to_keep]
filtered_data
```

```python
columns_to_plot = [
    'Altitude (m)', 'Vertical velocity (m/s)', 'Vertical acceleration (m/s²)',
    'Total velocity (m/s)', 'Total acceleration (m/s²)', 
    'Position East of launch (m)', 'Position North of launch (m)', 
    'Pitch rate (rad/s)', 'Yaw rate (rad/s)'
]

for column in columns_to_plot:
    plt.figure(figsize=(8, 5))
    plt.plot(filtered_data['Time (s)'], filtered_data[column], label=column)
    plt.title(f'{column} over Time')
    plt.xlabel('Time (s)')
    plt.ylabel(column)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```

```python
max_alt_time = filtered_data.loc[filtered_data['Altitude (m)'].idxmax(), 'Time (s)']
max_alt_value = filtered_data['Altitude (m)'].max()

max_alt_time, max_alt_value
```

```python
# Create a plot for each variable and add a vertical line at the time of maximum altitude for all plots
for column in columns_to_plot:
    plt.figure(figsize=(8, 5))
    plt.plot(filtered_data['Time (s)'], filtered_data[column], label=column)
    
    # Add a vertical line for the time of maximum altitude to all plots
    plt.axvline(x=max_alt_time, color='r', linestyle='--', label='Max Altitude Time')
    
    # For the altitude plot, also mark the max altitude point
    if column == 'Altitude (m)':
        plt.scatter(max_alt_time, max_alt_value, color='red', label='Max Altitude')

    plt.title(f'{column} over Time')
    plt.xlabel('Time (s)')
    plt.ylabel(column)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```

```python
# Focus on the time window closer to the maximum altitude
time_window = filtered_data[(filtered_data['Time (s)'] < max_alt_time) & (filtered_data['Time (s)'] > (max_alt_time - 10))]

# Looking for significant drops in velocity or acceleration, indicating air brake activation
velocity_diff_near_peak = time_window['Vertical velocity (m/s)'].diff().abs()
acceleration_diff_near_peak = time_window['Vertical acceleration (m/s²)'].diff().abs()

# Determine the threshold for significant changes in the near-peak region
threshold_velocity_near_peak = velocity_diff_near_peak.quantile(0.9)  # Top 10% of changes
threshold_acceleration_near_peak = acceleration_diff_near_peak.quantile(0.9)

# Find the time when these significant changes occur before the maximum altitude
air_brake_time_velocity_near_peak = time_window['Time (s)'][velocity_diff_near_peak > threshold_velocity_near_peak].min()
air_brake_time_acceleration_near_peak = time_window['Time (s)'][acceleration_diff_near_peak > threshold_acceleration_near_peak].min()

air_brake_time_velocity_near_peak, air_brake_time_acceleration_near_peak
```

```python
# Plot the vertical velocity and vertical acceleration again, now highlighting the refined air brake activation point

plt.figure(figsize=(8, 5))
# Plot Vertical Velocity
plt.plot(filtered_data['Time (s)'], filtered_data['Vertical velocity (m/s)'], label='Vertical velocity (m/s)')
# Highlight the air brake activation point based on the refined analysis
plt.axvline(x=air_brake_time_velocity_near_peak, color='g', linestyle='--', label='Air Brake Activation', linewidth=2)

plt.title('Vertical Velocity Over Time with Air Brake Activation')
plt.xlabel('Time (s)')
plt.ylabel('Vertical velocity (m/s)')
plt.grid(True)
plt.legend(loc='best')
plt.show()

plt.figure(figsize=(8, 5))
# Plot Vertical Acceleration
plt.plot(filtered_data['Time (s)'], filtered_data['Vertical acceleration (m/s²)'], label='Vertical acceleration (m/s²)')
# Highlight the air brake activation point based on the refined analysis
plt.axvline(x=air_brake_time_acceleration_near_peak, color='g', linestyle='--', label='Air Brake Activation', linewidth=2)

plt.title('Vertical Acceleration Over Time with Air Brake Activation')
plt.xlabel('Time (s)')
plt.ylabel('Vertical acceleration (m/s²)')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
for column in columns_to_plot:
    plt.figure(figsize=(8, 5))
    plt.plot(filtered_data['Time (s)'], filtered_data[column], label=column)
    
    plt.axvline(x=max_alt_time, color='r', linestyle='--', label='Max Altitude Time', linewidth=2)
    plt.axvline(x=air_brake_time_velocity_near_peak, color='g', linestyle='--', label='Air Brake Activation', linewidth=2)
    
    plt.title(f'{column} over Time with Max Altitude and Air Brake Activation')
    plt.xlabel('Time (s)')
    plt.ylabel(column)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```

```python
start_to_max_altitude = filtered_data[filtered_data['Time (s)'] <= max_alt_time]

for column in columns_to_plot:
    plt.figure(figsize=(8, 5))
    plt.plot(start_to_max_altitude['Time (s)'], start_to_max_altitude[column], label=column)
    
    plt.axvline(x=air_brake_time_velocity_near_peak, color='g', linestyle='--', label='Air Brake Activation', linewidth=2)
    plt.axvline(x=max_alt_time, color='r', linestyle='--', label='Max Altitude', linewidth=2)
    
    plt.title(f'{column} from Start to Max Altitude with Air Brake Activation')
    plt.xlabel('Time (s)')
    plt.ylabel(column)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()
```
